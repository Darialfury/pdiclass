```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img=cv2.imread('./images/oscura.png')
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV_FULL)
```

```python
plt.imshow(img_hsv[:,:,2], cmap=plt.cm.gray, vmin=0, vmax=255)
```

```python
imgv = img_hsv[:,:,2]
histogram, bins = np.histogram(imgv,bins=256,range=(0, 255))
```

```python
T = 255.*(np.cumsum(histogram))/imgv.size
imgv_eq = cv2.LUT(imgv,T)
```

```python
img_enhanced = img_hsv
img_enhanced[:,:,2] = imgv_eq
img_enhanced_rgb = cv2.cvtColor(img_enhanced, cv2.COLOR_HSV2BGR_FULL)

plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img_enhanced_rgb)
```
