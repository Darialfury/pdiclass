```python
#Functions to use
def cut_image(image_to_cut, points_bound):
    h, w, _ = image_to_cut.shape
    fp = np.array(points_bound[0])
    sp = np.array(points_bound[1])
    positive_image = image_to_cut[fp[1]:sp[1],fp[0]:sp[0]]
    disy, disx, _ = positive_image.shape
    
    mat_extract = np.zeros([h,w])
    mat_extract[fp[1]:fp[1]+disy,fp[0]:fp[0]+disx] = 255
    list_negative_image = []
    for i in range(0,4):
        flag = True
        while flag:
            a = np.random.randint(len(np.arange(0,h)))
            b = np.random.randint(len(np.arange(0,w)))
            if((a+disy)<h)and((b+disx)<w):
                box_extracted = mat_extract[a:a+disy,b:b+disx]
                if(np.sum(box_extracted)==0):
                    mat_extract[a:a+disy,b:b+disx] = 100 + (i*25)
                    neg_image = image_to_cut[a:a+disy,b:b+disx].copy()
                    list_negative_image.append(neg_image)
                    flag = False
    
    return positive_image, list_negative_image

#grid input image into cells
def grid_image(image_to_grid):
    h, w = image_to_grid.shape
    x = range(0,4)
    cellx = int(w/4)
    celly = int(h/4)
    join_cells = [image_to_grid[celly*j:celly*(j+1),cellx*(i):cellx*(i+1)] for i,j in itertools.product(x, repeat=2)]
    return join_cells

#Manual function to extract HOG features
def hog_manual(img, bin_n = 8):
    gx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
    gy = cv2.Sobel(img, cv2.CV_32F, 0, 1)
    mag, ang = cv2.cartToPolar(gx, gy)

    # quantizing binvalues in bin_n
    bins = np.int32(bin_n*ang/(2*np.pi))

    # Divide to 4 sub-squares
    bin_cells = bins[:10,:10], bins[10:,:10], bins[:10,10:], bins[10:,10:]
    mag_cells = mag[:10,:10], mag[10:,:10], mag[:10,10:], mag[10:,10:]
    hists = [np.bincount(b.ravel(), m.ravel(), bin_n) for b, m in zip(bin_cells, mag_cells)]
    hist = np.hstack(hists)
    return hist

def convert_gray_grid(list_neg):
    list_gray_cell = []
    for color_box in list_neg:
        neg_analyze = grid_image(cv2.cvtColor(color_box,cv2.COLOR_BGR2GRAY))
        list_gray_cell.append(neg_analyze)
    return list_gray_cell

def extract_negative_features(list_of_cells):
    set_of_hog = []
    for cells in list_of_cells:
        for each_cell in cells:
            set_of_hog.append(hog_manual(pos_cell[i]))
    return set_of_hog
```
