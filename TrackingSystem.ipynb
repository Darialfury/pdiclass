##### Show video in frames

```python
#Show frame by frame a video specified
dataset_name = './BenchmarkTracking/'
video_name = 'Deer'

path_video = os.path.join(dataset_name, video_name, 'img')
frame_list = os.listdir(path_video)
frame_list.sort()

for frame in frame_list:
    img_name = os.path.join(dataset_name,video_name,'img',frame)
    img_mat = cv2.imread(img_name)
    plt.imshow(img_mat)
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.002)
print "Finished..."
display.clear_output(wait=True)
```
