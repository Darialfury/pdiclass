```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Spatial filtering

```python
img_gray = cv2.imread("./images/ckt_board_saltpep_prob_pt05.png", 
                      cv2.CV_LOAD_IMAGE_GRAYSCALE) 
```

```python
kernel_size = 3
kernel = np.ones((kernel_size,kernel_size))/(kernel_size*kernel_size)
print kernel
```

```python
img_res = cv2.filter2D(img_gray,cv2.CV_8U,kernel)
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.imshow(img_res, cmap=plt.cm.gray, vmin=0, vmax=255)
```

```python
img_med = cv2.medianBlur(img_gray,3)
```

```python
plt.imshow(img_med,cmap=plt.cm.gray, vmin=0, vmax=255)
```

```python
kernel_laplacian = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
print kernel_laplacian
```

```python
moonlight = cv2.imread('./images/blurry_moon.png',cv2.CV_LOAD_IMAGE_GRAYSCALE)
sharpened = cv2.filter2D(moonlight,cv2.CV_8U,kernel_laplacian)
```

```python
plt.subplot(1,2,1)
plt.imshow(moonlight, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.subplot(1,2,2)
plt.imshow(sharpened, cmap=plt.cm.gray, vmin=0, vmax=255)
```

```python
sobel_y=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
print sobel_y
```

```python
sobel_x=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
print sobel_x
```

```python
gy = cv2.filter2D(moonlight,cv2.CV_32F,sobel_y)
gx = cv2.filter2D(moonlight,cv2.CV_32F,sobel_x)
```

```python
def compare(uno, dos, scale=True, cmap=plt.cm.gray):
    plt.subplot(1,2,1)
    
    kwargs = {"cmap": cmap}
    
    if scale:
        kwargs["vmin"] = 0
        kwargs["vmax"] = 255

    plt.imshow(uno, **kwargs)
    plt.subplot(1,2,2)
    plt.imshow(dos, **kwargs)
```

```python
compare(gx, gy, scale=False)
```

```python
compare(gx, gy)
```

```python
magG = np.sqrt(gx*gx + gy*gy)
plt.imshow(magG,cmap=plt.cm.gray, vmin=0, vmax=255)
```
