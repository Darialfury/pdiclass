```python
#Functions to use

# get all the lines of txt file and save them in a list
def get_items_txt(txt_name):
    lines_txt = open(txt_name, 'r')
    list_text = []
    for line in lines_txt:
        newline = line.split('\n')[0]
        list_text.append(newline)
    return list_text

#save and object
def save_object(obj, filename):
    with open(filename, 'wb') as output:
        pickle.dump(obj, output, -1)

#load an object
def load_object(filename):
    with open(filename, 'rb') as input:
        obj = pickle.load(input)
    return obj

#save and object
def save_list(my_list, filename):
    pickle.dump(my_list, open(filename, "wb"))

#load an object
def load_list(filename):
    keypoints_database = pickle.load( open( filename, "rb" ) )

def compute_keypoints((folder, path_dataset, newpath)):
    path_folder = path_dataset + folder + '/'
    list_images = os.listdir(path_folder)
    list_images.sort()
    path2_folder = newpath + folder
    if not os.path.exists(path2_folder):
        os.makedirs(path2_folder)
        
        # search for each image of the folder and compute sift
        for image in list_images:
            path_image_load = path_folder + '/' + image
            x = cv2.imread(path_image_load)
            y = cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
            sift = cv2.SIFT()
            kp, des = sift.detectAndCompute(y,None)

            # If descriptor are none, don't save it
            if des is not None:
                x2 = str(image.split('.',1)[0])
                path2 = path2_folder + '/' + x2 + '.txt'
                save_list(kp, path2)   
```

### Keypoint extraction

Extraction of keypoints of each image from dataset

```python
# Parameters of codebook to load
scale =300
k_clusters = 300
percentage_test = 0.30
```

```python
# save some centers of codebook
if scale != 'original':
    path_features = './images_sift_300/'
    path_dataset = './resize_images_dataset_300/'
    lstring_folders = os.listdir(path_features)
else:
    path_features = './images_sift/'
    lstring_folders = os.listdir(path_features)
    path_dataset = './101_ObjectCategories/'
path = './experiment_' + str(int(percentage_test*100))+ '_c' + str(k_clusters) + '_scale_' + str(scale) + '/code_book.pkl'
k_means = load_object(path)
codebook = k_means.cluster_centers_
some_centers = np.empty(shape=[0,128])
some_centers = np.vstack([some_centers, codebook[7][:], codebook[12][:], codebook[25][:], codebook[250][:]])
num_centers = some_centers.shape[0]
temp_distance = np.zeros(shape=[num_centers,1])
min_distance = np.zeros(shape=[num_centers,1])
temp_index_patch = np.zeros(shape=[num_centers,1])
index_patch = np.zeros(shape=[num_centers,1])
```

```python
# search patchs close to those some_centers
count = 0
print 'searching... '
for folder in lstring_folders:
    path_images = os.path.join(path_features, folder)
    list_sift_images = os.listdir(path_images)
    for image in list_sift_images:
        path_descriptor_load = os.path.join(path_images, image)
        image_use = image.split('.')[0] + '.jpg'
        path_in_dataset = os.path.join(path_dataset, folder, image_use)
        var = sio.loadmat(path_descriptor_load)
        array_to_stack = var['s_desc']
        for i in range(0,num_centers):
            temp_distance[i] = np.min(np.linalg.norm(array_to_stack - some_centers[i], axis = 1))
            temp_index_patch[i] = np.argmin(np.linalg.norm(array_to_stack - some_centers[i], axis = 1))
        if count != 0:
            condition = np.sum(temp_distance < min_distance)
            if condition != 0:
                index_patch[temp_distance < min_distance] = temp_index_patch[temp_distance < min_distance]
                tuple_index = np.where(temp_distance < min_distance)[0]
                min_distance[temp_distance < min_distance] = temp_distance[temp_distance < min_distance]
                for x in tuple_index:
                    images_patch[x] = path_in_dataset
        else:
            min_distance = np.array(temp_distance)
            index_patch = np.array(temp_index_patch)
            images_patch = [path_descriptor_load]*4 
        count += 1
```

```python
#show the patches closets to the centers
sift = cv2.SIFT()
for k in range(0,len(images_patch)):
    x = cv2.imread(images_patch[k])
    y = cv2.cvtColor(x,cv2.COLOR_BGR2GRAY)
    kp = sift.detect(y,None)
    px = int(kp[int(index_patch[k])].pt[0])
    py = int(kp[int(index_patch[k])].pt[1])
    patch = y[px-30:px+30,py-30:py+30]
    plt.subplot(1, num_centers, k+1)
    plt.imshow(patch,cmap=plt.cm.gray, vmin=0, vmax=255)
```
